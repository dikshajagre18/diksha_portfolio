```python
# replacing na values in college with No college
ocd_df["Previous Diagnoses"].fillna('', inplace = True)
ocd_df["Medications"].fillna('', inplace = True)
```

```python
# Group by 'Gender', calculate patient count and average (rounded to 2 decimal places)
grouped_stats_gender = new_ocd_df.groupby('Gender').agg(
    patient_count=('Patient ID', 'count'),
    avg_ybocs_obsessions=('Y-BOCS Score (Obsessions)', 'mean')
).round(2)

print(grouped_stats_gender)
grouped_stats_gender.to_csv('gender.csv')
```

```python
# Group by 'Ethnicity', calculate patient count and average (rounded to 2 decimal places)
grouped_stats_ethnicity = new_ocd_df.groupby('Ethnicity').agg(
    patient_count=('Patient ID', 'count'),
    avg_ybocs_obsessions=('Y-BOCS Score (Obsessions)', 'mean')
).round(2).sort_values(by='patient_count', ascending=False)

print(grouped_stats_ethnicity)
grouped_stats_ethnicity.to_csv('ethnicity.csv')
```

```python
new_ocd_df['Month'] = pd.DatetimeIndex(new_ocd_df['OCD Diagnosis Date']).month
```

```python
# Assuming 'OCD Diagnosis Date' is in datetime format, if not, convert it
new_ocd_df['OCD Diagnosis Date'] = pd.to_datetime(new_ocd_df['OCD Diagnosis Date'])

# Extract 'Month' and 'Year' from 'OCD Diagnosis Date'
new_ocd_df['Month'] = new_ocd_df['OCD Diagnosis Date'].dt.month
new_ocd_df['Year'] = new_ocd_df['OCD Diagnosis Date'].dt.year

# Get the list of current columns
columns = list(new_ocd_df.columns)

# Find the index of 'OCD Diagnosis Date' column
ocd_index = columns.index('OCD Diagnosis Date')

# Remove 'Month' and 'Year' from their current positions
columns.remove('Month')
columns.remove('Year')

# Insert 'Month' and 'Year' columns right after 'OCD Diagnosis Date'
columns.insert(ocd_index + 1, 'Month')
columns.insert(ocd_index + 2, 'Year')

# Reorder the DataFrame with the updated column order
new_ocd_df = new_ocd_df[columns]

# Now, the 'Month' and 'Year' columns will appear next to the 'OCD Diagnosis Date' column
new_ocd_df

```

```python
#Number of people diagnosed with OCD MoM
grouped_stats_month = new_ocd_df.groupby(['Year']).agg(
   patient_count=('Patient ID', 'count')
).sort_values(by='patient_count', ascending=False)

print(grouped_stats_month)
grouped_stats_month.to_csv('diagnosed with Month by Month.csv')
```

```python
#What is the most common Obsession Type (Count) & it's respective Average Obsession Score
grouped_stats_obsessiontype = new_ocd_df.groupby('Obsession Type').agg(
    patient_count=('Patient ID', 'count'),
    avg_ybocs_obsessions=('Y-BOCS Score (Obsessions)', 'mean')
).round(2).sort_values(by='patient_count', ascending=False)

print(grouped_stats_obsessiontype)
grouped_stats_obsessiontype.to_csv('obsessiontype.csv')
```

```python
#What is the most common Compulsion type (Count) & it's respective Average Obsession Score
grouped_stats_compulsiontype = new_ocd_df.groupby('Compulsion Type').agg(
    patient_count=('Patient ID', 'count'),
    avg_ybocs_obsessions=('Y-BOCS Score (Obsessions)', 'mean')
).round(2).sort_values(by='patient_count', ascending=False)

print(grouped_stats_compulsiontype)
grouped_stats_compulsiontype.to_csv('compulsiontype.csv')
```

```python
grouped_stats_anxietydiagnosis = new_ocd_df.groupby(['Age']).agg(
    anxiety_diagnosis_count = ('Anxiety Diagnosis', 'count'),
).sort_values(by='anxiety_diagnosis_count',ascending=False)

print(grouped_stats_anxietydiagnosis)
grouped_stats_anxietydiagnosis.to_csv('anxietydiagnosis.csv')
```

```python
grouped_stats_anxietydiagnosis_marital = new_ocd_df.groupby(['Marital Status']).agg(
    anxiety_diagnosis_count = ('Anxiety Diagnosis', 'count'),
).sort_values(by='anxiety_diagnosis_count',ascending=False)

print(grouped_stats_anxietydiagnosis_marital)
grouped_stats_anxietydiagnosis_marital.to_csv('anxietydiagnosis.csv')
```

```python
# Group by 'Education Level' and aggregate with unique values
grouped_stats_educationlevel = new_ocd_df.groupby('Education Level').agg(
    previous_diagnosis=('Previous Diagnoses', 'unique'),
    medications=('Medications', 'unique')
)

print(grouped_stats_educationlevel)

grouped_stats_educationlevel.to_csv('anxietydiagnosis.csv')
```
